# Linear Regression and Error Curve Plot

```python
import pandas as pd
import numpy as np
from __future__ import division
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read Data

```python
df = pd.read_csv("../../Recursos/regLin.csv")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df[["X"]],df[[df.columns[1]]], train_size=0.75)
```

```python
plt.scatter(X_train, Y_train)
```

## Linear Regression

```python
linReg = LinearRegression()
linReg.fit(X_train,Y_train)
print linReg.intercept_,linReg.coef_
np.mean((linReg.predict(X_test) - Y_test) **2)
```

## Prediction

```python
Y_pred = linReg.predict(X_test)
plt.scatter(X_test, Y_test,  color='blue')
plt.plot(X_test, Y_pred, color='red', linewidth=3)
```

## Calculating the error

```python
def calcError(X,y,w,w0):
    return np.mean(((X.dot(w) + w0)[0]-y[y.columns[0]])**2)
```

```python
print calcError(X_test, Y_test, linReg.coef_, linReg.intercept_)
print calcError(X_train, Y_train, linReg.coef_, linReg.intercept_)
```

```python
yt_arr = []
w_arr = []
for x in range (190,350):
    w = x/100 
    w_arr.append(w)
    yt_arr.append(calcError(X_train, Y_train, linReg.coef_, w))
```

```python
plt.plot(w_arr, yt_arr, color='red', linewidth=2)
plt.scatter(linReg.intercept_, calcError(X_train, Y_train, linReg.coef_, linReg.intercept_),  color='blue')
```
